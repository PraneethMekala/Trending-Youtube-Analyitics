```python
#create a list of IDs from data
df_us_list = df_us['video_id'].to_list()
```

```python
#adds rest of url to the id such that we can click and view video
x = 'https://www.youtube.com/watch?v='
us_urls = [ x + url for url in df_us_list]
us_urls[0]
```

```python
video_duration = []
```

```python
#if video unable to load, then assume 0 seconds and add to new list
#takes a long time to run! 
for url in us_urls:
    try:
        video = pafy.new(url).length
        video_duration.append(video)
        #df_us['duration'] = video_duration
        print(video,',',len(video_duration))
    except:
        video_duration.append(0)
        #df_us['duration'] = video_duration
        print(len(video_duration))
        
```

```python
#place all video durations into the dataframe as a new column
df_us['videoDuration'] = video_duration
```

```python
#export dataframe as csv
df_us.to_csv('USnvideos.csv', index=False)
```

```python
#used to grab data from websites
import urllib.request
```

```python
def url_to_jpg(i, url, file_path):
    filename = 'image-{}.jpg'.format(i)
    full_path = '{}{}'.format(file_path,filename)
    urllib.request.urlretrieve(url, full_path)
    
    print('{} saved.'.format(filename))
    return None
```
